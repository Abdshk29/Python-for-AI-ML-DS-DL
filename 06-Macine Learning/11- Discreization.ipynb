# Data Discretization
Data discretization is the process of converting continuous data into discrete data. In other words, it involves dividing continuous data into distinct, non-overlapping intervals or categories.



```python
#Import libraries
import pandas as pd
import seaborn as sns
from sklearn.preprocessing import KBinsDiscretizer
```

```python
df = sns.load_dataset('titanic')
df.head()
```

```python
# Impute missing values
df['age'] = df['age'].fillna(df['age'].median())
df['fare'] = df['fare'].fillna(df['fare'].median())
```

```python
# Discreatization of age

age_discretizer = KBinsDiscretizer(n_bins=2, encode='ordinal', strategy='uniform')
df['age_bin'] = age_discretizer.fit_transform(df[['age']])
df.head()
```

```python
sns.histplot(df['age'], kde=True)
```

```python
sns.histplot(df, x = 'age', hue='age_bin', kde=True)
```

```python
# Pandas for manual binning
df['age_bin2'] = pd.cut(df['age'], bins = 3, labels = [0, 1, 2])
sns.histplot(df, x = 'age', hue = 'age_bin2')
```

```python
# Define the bins 
bins = [0, 10, 18, 30,  50, 80]
labels = [ 'child', 'teenager', 'young adult', 'adult', 'old age']
df['age_bin3'] = pd.cut(df['age'], bins=bins, labels=labels)
sns.histplot(df, x = 'age', hue='age_bin3')
```
