```python
def play_game():
    board = [" " for _ in range(9)]
    current_player = "X"

    while True:
        display_board(board)
        print(f"Player {current_player}'s turn.")
        move = int(input("Enter a number (1-9) to place your mark: "))
        if board[move-1] == " ":
            board[move-1] = current_player
            if check_winner(board, current_player):
                display_board(board)
                print(f"Player {current_player} wins!")
                return
            if " " not in board:
                display_board(board)
                print("It's a tie!")
                return
            if current_player == "X":
                current_player = "O"
            else:
                current_player = "X"
        else:
            print("That spot is already taken. Try again.")

def display_board(board):
    print(f" {board[0]} | {board[1]} | {board[2]} ")
    print("---+---+---")
    print(f" {board[3]} | {board[4]} | {board[5]} ")
    print("---+---+---")
    print(f" {board[6]} | {board[7]} | {board[8]} ")

def check_winner(board, player):
    # Check rows
    for i in range(0, 9, 3):
        if set(board[i:i+3]) == {player}:
            return True
    # Check columns
    for i in range(3):
        if set([board[i], board[i+3], board[i+6]]) == {player}:
            return True
    # Check diagonals
    if set([board[0], board[4], board[8]]) == {player}:
        return True
    if set([board[2], board[4], board[6]]) == {player}:
        return True
    return False

play_game()
```

```python
import random

def play_hangman():
    # List of words to choose from
    words = ["python", "javascript", "ruby", "java", "csharp", "swift", "kotlin"]
    word = random.choice(words)
    
    # Initialize the game state
    guesses = 6
    correct_letters = set()
    incorrect_letters = set()
    
    # Game loop
    while True:
        # Display the current state of the game
        display_word(word, correct_letters)
        print(f"Incorrect letters: {', '.join(incorrect_letters)}")
        print(f"Guesses remaining: {guesses}")
        
        # Get a guess from the player
        guess = input("Guess a letter: ").lower()
        
        # Check if the guess is valid
        if len(guess) != 1 or not guess.isalpha():
            print("Please enter a single letter.")
            continue
        
        # Check if the letter has been guessed before
        if guess in correct_letters or guess in incorrect_letters:
            print("You already guessed that letter. Try again.")
            continue
        
        # Check if the letter is in the word
        if guess in word:
            print("Correct!")
            correct_letters.add(guess)
        else:
            print("Incorrect.")
            incorrect_letters.add(guess)
            guesses -= 1
        
        # Check if the player has won or lost
        if len(correct_letters) == len(set(word)):
            display_word(word, correct_letters)
            print("You win!")
            return
        if guesses == 0:
            print(f"You lose! The word was '{word}'.")
            return

def display_word(word, correct_letters):
    for letter in word:
        if letter in correct_letters:
            print(letter, end=" ")
        else:
            print("_", end=" ")
    print()

play_hangman()
```

```python
import random

# List of possible words
words = ['apple', 'banana', 'cherry', 'date', 'grape']

def play_wordle():
    # Choose a random word
    word = random.choice(words)
    
    # Give the player 6 attempts to guess the word
    for attempt in range(1, 7):
        # Get the player's guess
        guess = input(f"Attempt {attempt}: ").lower()
        
        # Check if the guess is valid
        if len(guess) != len(word):
            print("Invalid guess. The word must be the same length as the answer.")
            continue
        
        # Check the letters in the guess
        result = []
        for i in range(len(word)):
            if guess[i] == word[i]:
                result.append('🟩')
            elif guess[i] in word:
                result.append('🟨')
            else:
                result.append('⬜')
        
        # Print the result
        print(''.join(result))
        
        # Check if the player has won
        if guess == word:
            print(f"Congratulations! You guessed the word in {attempt} attempt(s).")
            return
    
    # If the player has used all 6 attempts and hasn't guessed the word, they lose
    print(f"Sorry, you didn't guess the word. The word was '{word}'.")

play_wordle()
```
